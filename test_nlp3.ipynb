```python
from scipy import sparse
from sklearn.feature_extraction.text import CountVectorizer
from string import punctuation
from collections import Counter
import pandas
import nltk
from nltk.stem.snowball import RussianStemmer
from nltk.tokenize import word_tokenize
```

```python
stemmer = RussianStemmer()
```

```python
text1 = []
greating1 = ['добр', 'здравств', 'приветств', 'привет']
for string in text:
    string1 = []
    for token in word_tokenize(string):
        if token not in punctuation:
            stem = stemmer.stem(token)
            string1.append(stem)
            s = ' '.join(string1)
    text1.append(s)
    
text1
```

```python
#фразы с приветствием
for w in greating1:
    for ww in text1:
        if w in ww:
            print(ww)
```

```python
#имя менеждера
import nltk
import pymorphy2
# probability score threshold
prob_thresh = 0.4
morph = pymorphy2.MorphAnalyzer()

for ww in text:
    for word in nltk.word_tokenize(ww):
        for p in morph.parse(word):
            if 'Name' in p.tag and p.score >= prob_thresh:
                print('{:<12}\t({:>12})\tscore:\t{:0.3}'.format(word, p.normal_form, p.score))
```

```python
#«В каждом диалоге обязательно необходимо поздороваться и попрощаться с клиентом»
greating1 = ['добр', 'здравств', 'приветств', 'привет']
farewell = ['доброго', 'свидания']
for i in manager.dlg_id.unique():
    manager_dlg = manager.loc[manager.dlg_id == i]
    text = []
    for row in manager_dlg['text']:
        text.append(row)
    text1 = []
    print('Диалог', i)
    for string in text:
        string1 = []
        for token in word_tokenize(string):
            if token not in punctuation:
                stem = stemmer.stem(token)
                string1.append(stem)
                s = ' '.join(string1)
        text1.append(s)
    #фразы с приветствием
    condition1 = 0
    condition2 = 0
    for w in greating1:
        for ww in text1:
            if w in ww:
                #print(ww)
                condition1 = 1
    #прощание
    for w in farewell:
        for ww in text:
            if w in ww:
                #print(ww)
                condition2 = 1
    #print(condition1, condition2)
    if condition1 == condition2:
        print('Менеджер в диалоге %i выполнил регламент' %i)
    else:
        print('Менеджер в диалоге %i НЕ выполнил регламент' %i)

```
